```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
X_train[2]
```

```python
X_train.shape
x_test.shape
```

```python
X_train=X_train/255
x_test = x_test/255
```

```python
# flatten the 2d data into 1d to give input to the input layer
X_train_flattened = X_train.reshape(len(X_train),28*28)
x_test_flattened = x_test.reshape(len(x_test),28*28)
```

```python
print(X_train_flattened.shape,x_test_flattened.shape)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
#model configuration
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
#perform on test dataset
model.evaluate(x_test_flattened,y_test)
```

```python
# plt.figure(figsize=(3,3))
plt.matshow(x_test[727])
```

```python
y_predicted = model.predict(x_test_flattened)
y_predicted[727]
```

```python
np.argmax(y_predicted[727])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
print(y_predicted_labels[:15])
print(y_test[:15])
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
#visualise the confusion matrix
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel("Truth")
```

```python
#adding hidden layers of 100 neurons
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid')
])
#model configuration
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
y_predicted[0]
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
print(y_predicted_labels[:15])
print(y_test[:15])
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel("Truth")
```
